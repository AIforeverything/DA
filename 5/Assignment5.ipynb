```python
#1.Comparison between various species based on sepal length and width using respective visualization
df1=df[['Species','SepalLengthCm','SepalWidthCm']].copy()
plt.bar(df1['Species'],df1['SepalLengthCm'],0.4)
plt.xlabel('Species')
plt.ylabel('SepalLength(cm)')
plt.xticks(rotation=45)
plt.show()
plt.bar(df1['Species'],df1['SepalWidthCm'],0.4)
plt.xlabel('Species')
plt.ylabel('SepalWidth(Cm)')
plt.xticks(rotation=45)
plt.show()

sns.scatterplot(data=df,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
```

```python
# 2.Perform the Bi-variate Analysis via pairplot visualization
sns.pairplot(df.drop(['Id'],axis=1),hue='Species',height=2)
```

```python
# 3.Justify Sepal Length and Sepal Width features are slightly correlated with each other
correlation=df1['SepalLengthCm'].corr(df1['SepalWidthCm'])
print("correlation: ",correlation)
plt.scatter(df1['SepalLengthCm'],df1['SepalWidthCm'])
print("Sepal Length and Sepal Width features are slightly correlated with each other.")
```

```python
# 4.Visualize the categorical feature “Species” with all other four input variables
df2=df.copy()
x=df2['Species']
y1=df2['SepalLengthCm']
y2=df2['SepalWidthCm']
y3=df2['PetalLengthCm']
y4=df2['PetalWidthCm']
# colors=('r','b','g','y')
fig,axes=plt.subplots(nrows=2,ncols=2)
ax0,ax1,ax2,ax3=axes.flatten()
ax0.bar(x,y1,0.4)
ax1.bar(x,y2,0.4)
ax2.bar(x,y3,0.4)
ax3.bar(x,y4,0.4)
```

```python
# 5.Plot the density of the length and width in the species.
df2=df.copy()
x=df2['Species']
y1=df2['SepalLengthCm']
y2=df2['SepalWidthCm']
y3=df2['PetalLengthCm']
y4=df2['PetalWidthCm']

fig,axes= plt.subplots(2,2,figsize=(6,6))
axes[0,0].hist(y1)
axes[0,0].set_title('SepalLengthCm')
axes[0,1].hist(y2)
axes[0,1].set_title(('SepalWidthCm'))
axes[1,0].hist(y3)
axes[1,0].set_title('PetalLengthCm')
axes[1,1].hist(y4)
axes[1,1].set_title('PetalWidthCm')
```

```python
# 6.Plot the probability density function(PDF) with each feature as a variable on X-axis and it’s histogram and corresponding kernel density plot on Y-axis
sns.displot(df.SepalLengthCm,kde=True)
sns.displot(df.SepalWidthCm,kde=True)
sns.displot(df.PetalLengthCm,kde=True)
sns.displot(df.PetalWidthCm,kde=True)
plt.show()
```
