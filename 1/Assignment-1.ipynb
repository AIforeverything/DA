```python
import pandas as pd
import numpy as np
df=pd.read_csv("raw_data_unmodified.csv")
df.head()
```

```python
df.size
```

```python
df.drop_duplicates(subset=['movie_title'],keep=False,inplace=True)
df.size
```

```python
df.reset_index(inplace=True,drop=True)
```

```python
df
```

```python
df.info()
```

```python
column_list=list(df.columns)
df['movie_title']=df['movie_title'].str.replace('?ÿ','')
df   
```

```python
# converting time columns (in this df 'year') into object datatype
column_list_refined=[]
time_list=[]
for item in column_list:
    if item.find('year')== -1:
        column_list_refined.append(item)
    if item.find('year')!= -1:
        time_list.append(item)
for item in time_list:
    df[item]=df[item].bfill()
    df[item]=df[item].astype(int)
    df[item]=df[item].astype(str)
```

```python
df['DIRECTOR_facebook_likes']=pd.to_numeric(df['DIRECTOR_facebook_likes'] , errors = 'coerce')
df['DIRECTOR_facebook_likes']
```

```python
#replacing null values of every numeric column with median
for item in column_list_refined:
     if ((df[item].dtype!=object) and (df[item].isnull)):  
            df.fillna(df[item].median(),inplace=True)  
```

```python
df.info()
df.to_csv("cleaned.csv")
```
