1. Most expensive car for each car company make.

2.Most expensive convertible vehicle for each company.

3.List all the cars which are automatic.

4. Premium cars of each car company

5. Average price of cars each car company make.

```python
import pandas as pd
import numpy as np
df=pd.read_csv("car.csv")
```

```python
df.head()
```

```python
df.groupby('Make')['MSRP'].max()
# 1.Most expensive car for each car company make.
```

```python
# 2.Most expensive convertible vehicle for each company.
df[df['Vehicle Style']=='Convertible']
```

```python
df[df['Vehicle Style']=='Convertible'].groupby('Make').MSRP.max()
```

```python
# 3.List all the cars which are automatic for each company make and model.
df[df['Transmission Type']=='AUTOMATIC'].groupby(['Make','Model']).first()
```

```python
# 4.Premium cars of each car company AND MODEL.
df[(df['Engine Fuel Type']=='premium unleaded (required)') | (df['Engine Fuel Type']=='premium unleaded (recommended)')].groupby(['Make','Model']).first()
```

```python
# 5.Average price of cars each car company make.
df.groupby('Make')['MSRP'].mean()
```
