```python
df1=df[['Company','Revenue (in millions)']].copy()
df2=df1.groupby('Company').sum('Revenue (in millions)').sort_values(by='Revenue (in millions)',ascending=False).head(5)
df2['Total_Revenue (in millions) of top 5 companies']=df2['Revenue (in millions)']
df2=df2.drop('Revenue (in millions)',axis=1)
df2['Total_Revenue (in millions) of top 5 companies'].astype(float)
print(df2)
plt.bar(df2.index,df2['Total_Revenue (in millions) of top 5 companies'])
plt.ylabel('Total_Revenue (in millions) of top 5 companies')
plt.xlabel('Company_Name')
plt.xticks(rotation=45)
plt.show()
```

Exxon Mobil total revenue is 5152459.6 millions which is greater than General Motors.

```python
df['Profit (in millions)']=pd.to_numeric(df['Profit (in millions)'],errors = 'coerce')
#converted df['Profit (in millions)'] to float dtype
```

```python
df.info()
```

```python
df2=df[['Company','Profit (in millions)']].copy()
df3=df2.groupby('Company').sum('Profit (in millions)').sort_values(by='Profit (in millions)',ascending=False).head(5)
df3['Total Profit (in millions) of top 5 companies']=df3['Profit (in millions)']
df3=df3.drop('Profit (in millions)',axis=1)
plt.bar(df3.index,df3['Total Profit (in millions) of top 5 companies'])
plt.ylabel('Total Profit (in millions) of top 5 companies')
plt.xlabel("Company_Name")
plt.xticks(rotation=45)
print(df3)
plt.show()
```

Exxon Mobil total profit is 403684.1 millions which is greater than General Motors (248127.3 millions ).
